# Problème général 

Dans les réseaux de télécommunications, attribuer des fréquences ou des canaux sans interférences est essentiel pour garantir une transmission efficace et sans perturbations entre les stations ou les antennes. Cela peut être modélisé comme un **problème de coloriage de graphe**.

Lorsque des stations de transmission ou des tours de télécommunication sont proches les unes des autres, leurs signaux peuvent interférer s'ils utilisent la même fréquence. Par conséquent, **on cherche à attribuer une fréquence différente aux stations adjacentes pour éviter ces interférences**.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import random
import networkx as nx
from geopy.distance import geodesic
```

```python
# Charger la carte du Bénin (shapefile ou GeoJSON)
url = "https://raw.githubusercontent.com/datasets/geo-boundaries-world-110m/master/countries.geojson"
world = gpd.read_file(url)

# Filtrer pour obtenir le Bénin uniquement
benin = world[world['admin'] == 'Benin']

# Fonction pour générer des points aléatoires dans la géométrie
def generate_random_points(polygon, num_points,seed=None):
    """
    Generate random points within a polygon.
    
    Args:
        polygon: A Shapely polygon within which to generate points.
        num_points: The number of points to generate.
        seed: An optional seed for reproducibility (default: None).
    
    Returns:
        A list of Shapely Point objects within the polygon.
    """
    if seed is not None:
        random.seed(seed)
    points = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(points) < num_points:
        # Générer un point aléatoire dans les limites
        random_point = Point(random.uniform(minx, maxx), random.uniform(miny, maxy))
        # Vérifier s'il est dans le polygone
        if polygon.contains(random_point):
            points.append(random_point)
    return points

# Générer 100 points aléatoires dans la surface du Bénin
num_points = 50
benin_polygon = benin.geometry.iloc[0]
random_points = generate_random_points(benin_polygon, num_points, seed=42)

# Créer un GeoDataFrame pour les points
points_gdf = gpd.GeoDataFrame(geometry=random_points, crs=benin.crs)

# Afficher la carte avec les points
fig, ax = plt.subplots(figsize=(10, 10))
benin.plot(ax=ax, color='lightblue', edgecolor='black')
points_gdf.plot(ax=ax, color='red', markersize=5)
plt.title("Carte du Bénin avec des points aléatoires pour le positionnement des antennes")
plt.show()
```

# Quelques définitions 

## 1. Nombre chromatique $\chi(G)$

Le **nombre chromatique** d'un graphe $G$ est le plus petit nombre de couleurs nécessaires pour colorier les sommets du graphe de sorte que deux sommets adjacents aient des couleurs différentes.

**Mathématiquement :**
$$
\chi(G) = \min \{ k \mid \text{le graphe } G \text{ est } k\text{-coloriage} \}.
$$

---

## 2. Coloration d'un graphe
Une **coloration d'un graphe** est l'attribution d'une couleur à chaque sommet ou arête selon des règles spécifiques :

- **Coloration des sommets** : Les sommets adjacents doivent avoir des couleurs différentes.
- **Coloration des arêtes** : Les arêtes incidentes au même sommet doivent avoir des couleurs différentes.

---

## 3. Graphe $k$-coloriable
Un graphe est dit **$k$-coloriable** s'il est possible d'attribuer $k$ couleurs ou moins à ses sommets de manière à ce que deux sommets adjacents n'aient jamais la même couleur.

---

## 4. Classe chromatique
La **classe chromatique** est liée à la coloration des arêtes. Elle représente le nombre minimum de couleurs nécessaires pour colorier les arêtes d'un graphe de manière à ce que deux arêtes incidentes au même sommet aient des couleurs différentes.

---

## 5. Problème chromatique
Le **problème chromatique** est un problème d'optimisation qui consiste à trouver le **nombre chromatique** $\chi(G)$ pour un graphe donné.  

---

## 6. Clique
Une **clique** dans un graphe est un sous-ensemble de sommets où chaque paire de sommets est connectée par une arête.  
Le **nombre de clique** $\omega(G)$ est la taille de la plus grande clique du graphe.

Dans un graphe, on a :
$$
\chi(G) \geq \omega(G),
$$
car les sommets d'une clique doivent tous avoir des couleurs différentes.

---

## 7. Delta ($\Delta$) - Degré maximum
Le degré maximum d'un graphe $\Delta(G)$ est le nombre maximum d'arêtes connectées à un sommet.

Il est souvent une borne supérieure pour $\chi(G)$, c'est-à-dire :
$$
\chi(G) \leq \Delta(G) + 1.
$$

```python
# Rayon prédéterminé en kilomètres
radius_km = 50

# Convertir les points en un graphe NetworkX
G = nx.Graph()

# Ajouter les points comme nœuds avec leurs coordonnées
for i, point in enumerate(random_points):
    G.add_node(i, coords=(point.y, point.x))  # Stocker en (lat, lon)

# Ajouter les arêtes si la distance entre les points est inférieure au rayon
for i in range(len(random_points)):
    for j in range(i + 1, len(random_points)):
        point_i = G.nodes[i]['coords']
        point_j = G.nodes[j]['coords']
        distance = geodesic(point_i, point_j).km
        if distance <= radius_km:
            G.add_edge(i, j, weight=distance)

# Afficher le graphe sur la carte
fig, ax = plt.subplots(figsize=(10, 10))
benin.plot(ax=ax, color='lightblue', edgecolor='black')

# Tracer les arêtes
for edge in G.edges:
    node1, node2 = edge
    point1 = random_points[node1]
    point2 = random_points[node2]
    plt.plot([point1.x, point2.x], [point1.y, point2.y], color='gray', linewidth=0.5)

# Tracer les sommets
points_gdf.plot(ax=ax, color='red', markersize=10)
plt.title(f"Graphe des stations avec des connexions dans un rayon de {radius_km} km")
plt.show()
```

# Formulation mathématique :

## 1. Données :
- Un graphe $G = (V, E)$, où $V$ est l'ensemble des sommets et $E$ l'ensemble des arêtes.
- Un ensemble de couleurs $C = \{1, 2, \dots, k\}$.

## 2. Variables :
- $x_{i,c} \in \{0, 1\} : x_{i,c} = 1$ si le sommet $i$ est colorié avec la couleur $c$, 0 sinon.

## 3. Contraintes :
- Chaque sommet $i \in V$ doit avoir exactement une couleur :
  $$
  \sum_{c \in C} x_{i,c} = 1 \quad \forall i \in V.
  $$
- Deux sommets adjacents $(i, j) \in E$ ne doivent pas avoir la même couleur :
  $$
  x_{i,c} + x_{j,c} \leq 1 \quad \forall (i, j) \in E, \, \forall c \in C.
  $$

Minimiser $k$, le nombre de couleurs nécessaires.

# Résolution
Le problème de coloriage est **NP-difficile**. Trouver le nombre chromatique minimal d'un graphe général est computationalement coûteux. Pour cette raison, des approches heuristiques et approximatives sont souvent utilisées.
## Algorithmes pour résoudre le problème

### 1. Algorithme Glouton
- Assigner une couleur à chaque sommet en respectant les contraintes, en parcourant les sommets dans un ordre déterminé (souvent décroissant en degré).
- Simple mais pas optimal.

### 2. Coloration par Recherche Locale
- Optimiser une solution initiale (par exemple, obtenue avec l'algorithme glouton) en réduisant le nombre de couleurs par des permutations locales.

### 3. Programmation par Contraintes
- Modéliser le problème comme un problème de satisfaction de contraintes (CSP) et utiliser des solveurs comme **ortools**.

### 4. Algorithme Exact (Branch and Bound)
- Utiliser des approches exactes pour trouver une solution optimale.
- Plus coûteux en temps.

### 5. Algorithme de Métaheuristiques
- Utiliser des approches comme les algorithmes génétiques, le recuit simulé ou les colonies de fourmis pour chercher des solutions proces de l'optimum.


# Implémentation de l'algorithme glouton (greedy algorithm): 

```python
# Algorithme glouton pour la coloration
def greedy_coloring(graph):
    colors = {}
    for node in sorted(graph.nodes, key=lambda x: len(graph[x]), reverse=True):  # Trier par degré
        neighbor_colors = {colors[neighbor] for neighbor in graph.neighbors(node) if neighbor in colors}
        for color in range(len(graph)):  # Essayer les couleurs disponibles
            if color not in neighbor_colors:
                colors[node] = color
                break
    return colors

# Colorer le graphe
coloring = greedy_coloring(G)
```

```python
# Afficher le graphe sur la carte du Bénin avec les couleurs
fig, ax = plt.subplots(figsize=(10, 10))

# Tracer la carte du Bénin
benin.plot(ax=ax, color='lightblue', edgecolor='black')

# Tracer les arêtes
for edge in G.edges:
    node1, node2 = edge
    point1 = random_points[node1]
    point2 = random_points[node2]
    plt.plot([point1.x, point2.x], [point1.y, point2.y], color='gray', linewidth=0.5)

# Tracer les nœuds avec leurs couleurs
for node, point in enumerate(random_points):
    plt.scatter(
        point.x, 
        point.y, 
        color=plt.cm.rainbow(coloring[node] / max(coloring.values())), 
        s=30, 
        label=f"Station {node}" if node == 0 else ""
    )

# Calculer le nombre total de couleurs utilisées
num_colors = len(set(coloring.values()))
print(f"Nombre total de couleurs utilisées : {num_colors}")

# Optionnel : afficher sur la figure
plt.text(
    0.5, -0.1,  # Position relative sur la figure
    f"Nombre total de couleurs utilisées : {num_colors}", 
    ha='center', 
    va='top', 
    transform=ax.transAxes, 
    fontsize=12, 
    bbox=dict(facecolor='white', alpha=0.8)
)

plt.title(f"Graphe des stations avec coloration des sommets sur la carte du Bénin")
plt.show()
```
